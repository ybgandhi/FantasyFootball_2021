```python
# import libraries
import os
import pandas as pd
```

```python
# variable for download date of file
download_date = "Aug_24"
```

```python
# load csv file
rankings_file = "../rankings/"+ download_date + ".csv"
rankings = pd.read_csv(rankings_file)
```

```python
rankings
```

```python
# create position column from POS column
rankings["Position"] = rankings["POS"].astype(str).str[0:2]
rankings
```

```python
# numeric value for Strength of Schedule
rankings["SOS SEASON"] = rankings["SOS SEASON"].astype(str).str[0:1]
rankings.to_csv("../analysis/full.csv")
rankings
```

```python
# reorder columns in dataframe 
preference = rankings[["PLAYER NAME","TEAM","RK","TIERS","SOS SEASON","BYE WEEK"]]
#preference.to_csv("../analysis/full.csv")
preference
```

```python
# set Running Back dataframe
running_backs = preference.loc[rankings['Position'] == "RB"]
running_backs = running_backs[running_backs["TIERS"]<11]
running_backs = running_backs.reset_index(drop=True)
running_backs.to_csv("../analysis/running_backs.csv")
```

```python
# set Wide Receiver dataframe
wide_receiver = preference.loc[rankings['Position'] == "WR"]
wide_receiver = wide_receiver[wide_receiver["TIERS"]<11]
wide_receiver = wide_receiver.reset_index(drop=True)
wide_receiver.to_csv("../analysis/wide_receiver.csv")
```

```python
# set Quarterback dataframe
quarterback = preference.loc[rankings['Position'] == "QB"]
quarterback = quarterback[quarterback["TIERS"]<9]
quarterback = quarterback.reset_index(drop=True)
quarterback.to_csv("../analysis/quarterback.csv")
```

```python
# set Tight End dataframe
tight_end = preference.loc[rankings['Position'] == "TE"]
tight_end = tight_end[tight_end["TIERS"]<13]
tight_end = tight_end.reset_index(drop=True)
tight_end.to_csv("../analysis/tight_end.csv")
```

```python
# set Defense/Special Teams dataframe
dst = preference.loc[rankings['Position'] == "DS"]
#dst = dst[dst["TIERS"]<6]
dst = dst.reset_index(drop=True)
dst.to_csv("../analysis/dst.csv")
```

```python
# set Kicker dataframe
kicker = preference.loc[rankings['Position'] == "KS"]
kicker = kicker[kicker["TIERS"]<4]
kicker = kicker.reset_index(drop=True)
kicker
#kicker.to_csv("../analysis/kicker.csv")
```
