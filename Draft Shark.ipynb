```python
import pandas as pd
```

```python
# Draft Sharks URL's

## half ppr ADP rankings
half_adp_url = "https://www.draftsharks.com/adp/half"
```

```python
adp_half_tables = pd.read_html(half_adp_url)
adp_half_tables
```

```python
adp_half_df = adp_half_tables[0]
adp_half_df
```

```python
count = adp_half_df.count
print(count)
```

```python
# set up dataframe 
adp_half_df["DF Rank"] = adp_half_df["Positional Rank"].astype(str).str[0:2]
adp_half_df["ADP Avg."] = adp_half_df["FF Nerds Avg. Draft Position"].astype(str).str[0:5]
adp_half_df["Position"] = adp_half_df["Positional Rank"].astype(str).str[-2:]
adp_half_df["Position"] = adp_half_df["Position"].replace({"EF":"DEF"})
adp_half_df["Positional Rank"] = adp_half_df["Positional Rank"].str.replace('\d+','')
adp_half_df["Positional Rank"] = adp_half_df["Positional Rank"].str.replace('. ','')
adp_half_df["Positional Rank"] = adp_half_df["Positional Rank"].str.replace('RB','')
adp_half_df["Positional Rank"] = adp_half_df["Positional Rank"].str.replace('WR','')
adp_half_df["Positional Rank"] = adp_half_df["Positional Rank"].str.replace('DEF','')
adp_half_df["Positional Rank"] = adp_half_df["Positional Rank"].str.replace('QB','')
adp_half_df["Positional Rank"] = adp_half_df["Positional Rank"].str.replace('TE','')
adp_half_df["Name"] = adp_half_df["Positional Rank"]
#adp_half_df = adp_half_df.loc[0:250:]

adp_half_df
```

```python
DS_Half_PPR = adp_half_df[["DF Rank","ADP Avg.","Name",
                           "Position","Bye Week","Consensus Points"]]
DS_Half_PPR = DS_Half_PPR.set_index("DF Rank")
DS_Half_PPR.to_csv("../FantasyFootball_2021/DS/half_ppr.csv")
```
